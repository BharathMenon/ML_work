```python
X_scaled = X/255
X_test_scaled = X_test/255
X_scaled.shape
```

```python
feature_extractor_model = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
pretrained_model_without_top_layer = hub.KerasLayer(feature_extractor_model,input_shape=(224,224,3),trainable=False)
base_model = tf.keras.applications.MobileNetV2(input_shape=(224,224,3),
                                               include_top=False,
                                               weights='imagenet')
base_model.trainable = False
import tensorflow as tf

# Create a dummy input tensor with shape (1, 224, 224, 3)
dummy_input = tf.zeros((1, 224, 224, 3))

# Pass the dummy input through the pretrained_model_without_top_layer
dummy_output = base_model(dummy_input)

# Get the output shape
output_shape = dummy_output.shape

print(output_shape)
```

```python

model = tf.keras.Sequential([
    base_model,
    #layers.Reshape((40,8,4)),
    #layers.Flatten(),
    #layers.GlobalAveragePooling2D(),
    layers.Conv2D(filters=20,kernel_size=(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.2),
    layers.Flatten(),
    layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='SGD',loss=tf.keras.losses.BinaryCrossentropy(),metrics=['accuracy'])
model.fit(X_scaled,y,epochs=10,batch_size=10)
```

```python
model.evaluate(X_test_scaled,y_test)
```
