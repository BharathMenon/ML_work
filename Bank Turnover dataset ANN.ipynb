```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv("C:/Users/Bharath/Desktop/archive (1)/Churn_Modelling.csv")
df.dtypes
```

```python
df
```

```python
for column in df:
    print(f' {column} :{df[column].unique()}')
```

```python
df['Gender'].replace({'Female':1,'Male':0},inplace=True)  
```

```python
df.drop('CustomerId',axis='columns',inplace=True)
```

```python
df
```

```python
df2 = pd.get_dummies(data=df,columns=['Surname','Geography'])
```

```python
cols_to_scale = ['CreditScore','Age','Balance','EstimatedSalary','NumOfProducts']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2
```

```python
X = df2.drop(['Exited'],axis='columns')
y = df2['Exited']
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
model = keras.Sequential([
    keras.layers.Dense(2500,input_shape=(2944,),activation='relu'),
    keras.layers.Dense(1200,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='SGD',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=10)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
ypred = []
for i in y_pred:
    if i<=0.5:
        ypred.append(0)
    else:
        ypred.append(1)
```
