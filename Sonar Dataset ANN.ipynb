```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("C:/Users/Bharath/Desktop/DL_notebooks/Datasets/sonar_dataset.csv",header=None)
df.sample(5)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
X = df.drop(60,axis = 'columns')
y = df[60]
```

```python
 y = pd.get_dummies(y,drop_first = True) #R-->1 and M-->0
 y    
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
model = keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,epochs = 100,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cr = classification_report(y_test,y_pred)
cr
```

```python
#Using fDropout layers
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
model = keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,epochs = 100,batch_size=10)
```

```python
model.evaluate(X_test,y_test)
```
